```python
#membuat bentuk geometri 2d dari open cv diawali dengan pembuatan black image

black_img = np.zeros((512,512,3), np.uint8)
plt.imshow(black_img)
```

```python
#perhatikan koordinat titik2 pt1 dan pt2
cv.rectangle(black_img,pt1=(384,0),pt2=(510,150),color=(0,255,0), thickness=10)
plt.imshow(black_img)
```

```python
cv.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,0,255), thickness=15)
plt.imshow(black_img)
```

tahap selanjutnya ditambah menambahkan bentuk lingkaran sesuai radius yang  di tulis  pada kode program

```python
cv.circle(black_img,center=(100,100),radius=50,color=(255,0,0), thickness=8)
plt.imshow(black_img)
```

kemudian dilakukan penambahan garis sesuai dengan koordinat pt1 dan pt2 sbb

```python
cv.line(black_img,pt1=(0,0),pt2=(512,512),color=(255,255,255), thickness=5)
plt.imshow(black_img)
```

penambahan text dengan font yang telah tertulis dengan  ukuran yang sudah titentukan

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,0),thickness=5,lineType=cv.LINE_AA)
plt.imshow(black_img)
```

```python
black_img2=np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(black_img2)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
vertices = vertices.reshape((-1,1,2)) #nilai 2 untuk menunjukkan bahwa tiap titik dibuat 3 chanel yang mewakili R, G ,dan B
print(vertices)
```

```python
cv.polylines(black_img2,[vertices],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(black_img2)
```
