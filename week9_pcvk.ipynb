Buatlah fungsi dilasi serta tampilkan hasil dilasi pada citra dengan Structuring Element
Dilasi (SED) berbentuk cross 3 x 3 tanpa menggunakan library morphology dari
openCV, sehingga menghasilkan tampilan seperti di bawah ini:

```python
#Dilasi tanpa library
def dilasi_citra(F,w):
    #size image
    p,q= F.shape
    imgD= np.zeros((p,q), dtype=np.uint8)
    #Generate structure element dilasi (SED)
    SED= np.ones((w,w), dtype=np.uint8)
    constant1= (w-1)//2
    #Proses Dilasi
    for i in range(constant1, p-constant1):
        for j in range(constant1,q-constant1):
            temp= F[i-constant1:i+constant1+1, j-constant1:j+constant1+1]
            product= temp*SED

            imgD[i,j]= np.max(product)
    return imgD
img_d= cv2.imread('/content/drive/MyDrive/pcvk/platnomer.jpg',0)
ret, thresh = cv2.threshold(img_d, 125, 200, cv2.THRESH_BINARY)
w=5
imgD = dilasi_citra(thresh,w)
plt.subplot(131),plt.imshow(img_d,cmap = 'gray')
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(imgD,cmap = 'gray')
plt.title('Citra Hasil Dilasi'), plt.xticks([]), plt.yticks([])
plt.show()
```

Buatlah fungsi erosi serta tampilkan hasil erosi pada citra dengan Structuring Element Erosi (SE) berbentuk square 5 x 5 tanpa menggunakan library morphology dari openCV, sehingga menghasilkan tampilan seperti di bawah ini:

```python
#Erosi Tanpa Library
def erosi_citra(F,k):
    #size image
    m,n= F.shape
    #Generate structure element erosi (SE)
    SE= np.ones((k,k), dtype=np.uint8)
    constant= (k-1)//2
    imgE= np.zeros((m,n), dtype=np.uint8)
    #Proses Erosi
    for i in range(constant, m-constant):
        for j in range(constant,n-constant):
            temp= F[i-constant:i+constant+1, j-constant:j+constant+1]
            product= temp*SE
            imgE[i,j]= np.min(product)
    return imgE
img_e= cv2.imread('/content/drive/MyDrive/pcvk/platnomer.jpg',0) #
ret1, thresh1 = cv2.threshold(img_e, 175, 225, cv2.THRESH_BINARY)
k=5
imgE = erosi_citra(thresh1,k)
plt.subplot(131),plt.imshow(img_e,cmap = 'gray')
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(imgE,cmap = 'gray')
plt.title('Citra Hasil Erosi'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Memuat citra KTP menggunakan path yang Anda berikan
path_ktp = "/content/drive/MyDrive/pcvk/KTP.jpg"
img = cv2.imread(path_ktp, cv2.IMREAD_GRAYSCALE)

# Memeriksa apakah citra berhasil dimuat
if img is None:
    print("Citra tidak ditemukan, pastikan path benar.")
else:
    # Menyempurnakan citra dengan thresholding untuk meningkatkan kontras
    _, threshold_img = cv2.threshold(img, 128, 255, cv2.THRESH_BINARY_INV)

    # Kernel untuk operasi morfologi
    kernel = np.ones((3, 3), np.uint8)

    # Melakukan berbagai operasi morfologi
    dilated = cv2.dilate(threshold_img, kernel, iterations=1)
    eroded = cv2.erode(threshold_img, kernel, iterations=1)
    opened = cv2.morphologyEx(threshold_img, cv2.MORPH_OPEN, kernel)
    closed = cv2.morphologyEx(threshold_img, cv2.MORPH_CLOSE, kernel)
    gradient = cv2.morphologyEx(threshold_img, cv2.MORPH_GRADIENT, kernel)
    tophat = cv2.morphologyEx(threshold_img, cv2.MORPH_TOPHAT, kernel)
    blackhat = cv2.morphologyEx(threshold_img, cv2.MORPH_BLACKHAT, kernel)

    # Menampilkan hasil semua operasi morfologi
    titles = [
        'Original', 'Thresholded', 'Dilated', 'Eroded',
        'Opened', 'Closed', 'Gradient', 'Tophat', 'Blackhat'
    ]
    images = [
        img, threshold_img, dilated, eroded,
        opened, closed, gradient, tophat, blackhat
    ]

    plt.figure(figsize=(15, 10))
    for i in range(9):
        plt.subplot(3, 3, i + 1)
        plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])
        plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Memuat citra KTP
path_ktp = "/content/drive/MyDrive/pcvk/KTP.jpg"
img = cv2.imread(path_ktp, cv2.IMREAD_GRAYSCALE)

# Memeriksa apakah citra berhasil dimuat
if img is None:
    print("Citra tidak ditemukan, pastikan path benar.")
else:
    # Threshold untuk membuat citra biner
    _, binary = cv2.threshold(img, 128, 255, cv2.THRESH_BINARY_INV)

    # Kernel untuk erosi
    kernel = np.ones((3, 3), np.uint8)
    skeleton = np.zeros(binary.shape, np.uint8)

    # Iterasi erosi dan dilasi untuk mendapatkan skeleton
    while True:
        eroded = cv2.erode(binary, kernel)
        temp = cv2.dilate(eroded, kernel)
        temp = cv2.subtract(binary, temp)
        skeleton = cv2.bitwise_or(skeleton, temp)
        binary = eroded.copy()

        # Hentikan jika tidak ada perubahan lagi
        if cv2.countNonZero(binary) == 0:
            break

    # Menampilkan hasil skeleton
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1), plt.imshow(img, cmap='gray')
    plt.title("Original Image"), plt.axis("off")
    plt.subplot(1, 2, 2), plt.imshow(skeleton, cmap='gray')
    plt.title("Skeletonized Image"), plt.axis("off")
    plt.tight_layout()
    plt.show()
```
