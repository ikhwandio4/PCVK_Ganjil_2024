```python
# Load gambar lena.jpg
img = Image.open('/content/drive/MyDrive/pcvk/lena.jpg').convert('RGB')

# Inisialisasi histogram
hist = [0] * 256

# Hitung jumlah kemunculan setiap pixel warna pada seluruh pixel citra yang ada
for i in range(img.size[0]):
    for j in range(img.size[1]):
        r, g, b = img.getpixel((i, j))
        hist[r] += 1
        hist[g] += 1
        hist[b] += 1

# Buat histogram dengan sumbu x adalah intensitas warna dan sumbu y jumlah kemunculan
print("Histogram:")
for i in range(256):
    print(f"Intensitas Warna {i}: {hist[i]}")

# Plot histogram
import matplotlib.pyplot as plt
plt.bar(range(256), hist)
plt.xlabel('Intensitas Warna')
plt.ylabel('Jumlah Kemunculan')
plt.title('Histogram Citra')
plt.show()
```

5. Buatlah histogram citra seperti output histogram equalization dan juga tampilan gambar sebelum dan sesudah proses histogram equalization berikut berdasarkan flowchart di bawah ini : (Gunakan gambar lena_lc.jpg)

```python


import numpy as np
import matplotlib.pyplot as plt
from skimage import io, img_as_float

# Load the input image
img = io.imread('/content/drive/MyDrive/pcvk/lena_lc.jpg')

# Count the occurrence of each color intensity value
hist, bins = np.histogram(img.ravel(), 256, [0, 256])

# Calculate the cumulative sum of frequencies
cum_sum = np.cumsum(hist)
cum_sum = cum_sum / cum_sum[-1]

# Perform histogram equalization
eq_img = np.zeros_like(img)
for i in range(256):
    eq_img[img == i] = (255 * cum_sum[i]).astype(np.uint8)

# Display the original and resulting images
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(img, cmap='gray')
axs[0].set_title('Original Image')

axs[1].imshow(eq_img, cmap='gray')
axs[1].set_title('Histogram Equalization Result')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the input image
img = cv2.imread('/content/drive/MyDrive/pcvk/lena_lc.jpg')

# Convert the image to grayscale (single-channel 8-bit unsigned integer)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Perform histogram equalization using OpenCV's equalizeHist function
eq_img = cv2.equalizeHist(gray_img)

# Display the original and resulting images
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original Image')

axs[1].imshow(eq_img, cmap='gray')
axs[1].set_title('Histogram Equalization Result')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load gambar lena.jpg
img = cv2.imread('/content/drive/MyDrive/pcvk/lena.jpg')

# Definisikan warna-warna yang akan digunakan
warna = [
    (0, 0, 0),  # Hitam
    (0, 255, 0),  # Hijau
    (255, 255, 0),  # Kuning
    (0, 0, 255),  # Biru
    (0, 255, 255),  # Cyan
    (255, 0, 255),  # Magenta
    (255, 255, 255),  # Putih
]

# Buat fungsi untuk melakukan pemetaan warna
def pemetaan_warna(pixel):
    jarak_min = float('inf')
    warna_terdekat = None
    for w in warna:
        jarak = np.linalg.norm(np.array(pixel) - np.array(w))
        if jarak < jarak_min:
            jarak_min = jarak
            warna_terdekat = w
    return warna_terdekat

# Lakukan pemetaan warna pada gambar
img_pemetaan = np.zeros_like(img)
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        img_pemetaan[i, j] = pemetaan_warna(img[i, j])

# Display hasil pemetaan
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Gambar Asli')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(img_pemetaan, cv2.COLOR_BGR2RGB))
plt.title('Hasil Pemetaan')

plt.show()
```
